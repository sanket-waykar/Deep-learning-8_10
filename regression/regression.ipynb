```python
from warnings import filterwarnings
filterwarnings ("ignore")
```

### Genrate the dataset for regression

```python
from sklearn.datasets import make_regression
X,Y = make_regression(n_samples=1000,n_features=10,n_informative=6,noise=10,random_state=42)
```

```python
X
```

```python
Y
```

```python
import pandas as pd
X=pd.DataFrame(X)
X
```

```python
cols=[f"x{i+1}"for i in range(10)]
cols
```

```python
X.columns=cols
X.head()
```

```python
Y=pd.DataFrame(Y)
Y.columns=["y"]
```

```python
Y
```

```python
X.isna().sum()
```

### Preprocessing on X Dataframe - Scikit Learn Pipelines

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())])
```

```python
X_pre =num_pipe.fit_transform(X)
X_pre
```

```python
cols = num_pipe.get_feature_names_out()
cols
```

```python
X_pre = pd.DataFrame(X_pre, columns=cols)
X_pre.head()
```

```python
X.mean()
```

```python
X_pre.mean()
```

```python
X_pre.std()
```

### Train Test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest= train_test_split(X_pre,Y,test_size=0.2,random_state=42)
```

```python
xtrain.head()
```

```python
ytrain
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Create a Keras Model

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
xtrain.shape[0]
```

```python
xtrain.shape[1]
```

```python
#intialize the neural network model
nn=Sequential()

#input + hidden layer 1
nn.add(Dense(64,activation='relu',input_dim=xtrain.shape[1]))

#hidden layer 2
nn.add(Dense(32,activation='relu'))

#output
nn.add(Dense(1,activation="linear"))
```

### Compile the Model

```python
nn.compile(optimizer="adam",loss='mse',metrics=['mae'])
```

```python
nn.summary()
```

### Fit the Neural Network

```python
hist=nn.fit(xtrain,ytrain,epochs=100,validation_split=0.2)
```

```python
hist.history['val_loss']
```

### Plot the learning curve

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(hist.history["loss"],label="Train msg")
plt.plot(hist.history["val_loss"],label="Validation MSE")
plt.xlabel("Epochs")
plt.ylabel("MSE")
plt.title("Learning Curve")
plt.legend()
plt.show()
```

### Evaluate the model

```python
nn.evaluate(xtrain,ytrain)
```

```python
nn.evaluate(xtest,ytest)
```

### predict train and Test

```python
ypred_tr = nn.predict(xtrain)
ypred_ts = nn.predict(xtest)
```

### R2 scores in train and test

```python
from sklearn.metrics import r2_score
tr_r2=r2_score(ytrain,ypred_tr )
tr_r2
```

```python
ts_r2=r2_score(ytest,ypred_ts)
ts_r2
```
